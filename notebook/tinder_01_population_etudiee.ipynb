Transformation de valeurs

```python
# Map des valeurs
gender_map = {
    0: "Female",
    1: "Male"
}

condtn_map = {
    1: "limited choice",
	2: "extensive choice"
}

yes_no_map = {
    0: "no",
    1: "yes"
}

field_code_map = {
    1: "Law",
    2: "Math",
    3: "Social Science, Psychologist",
    4: "Medical Science, Pharmaceuticals, and Bio Tech",
    5: "Engineering",
    6: "English/Creative Writing/ Journalism",
    7: "History/Religion/Philosophy",
    8: "Business/Econ/Finance",
    9: "Education, Academia",
    10: "Biological Sciences/Chemistry/Physics",
    11: "Social Work",
    12: "Undergrad/undecided",
    13: "Political Science/International Affairs",
    14: "Film",
    15: "Fine Arts/Arts Administration",
    16: "Languages",
    17: "Architecture",
    18: "Other"
}

race_map = {
    1: "Black/African American",
	2: "European/Caucasian-American",
	3: "Latino/Hispanic American",
	4: "Asian/Pacific Islander/Asian-American",
	5: "Native American",
	6: "Other"
}

goal_map = {
    1: "Seemed like a fun night out",
	2: "To meet new people",
	3: "To get a date",
	4: "Looking for a serious relationship",
	5: "To say I did it",
	6: "Other"
}

date_go_out_map = {
    1: "Several times a week",
	2: "Twice a week",
	3: "Once a week",
	4: "Twice a month",
	5: "Once a month",
	6: "Several times a year",
	7: "Almost never"
}

career_code_map = {
    1: "Lawyer",
    2: "Academic/Research",
    3: "Psychologist",
    4: "Doctor/Medicine",
    5: "Engineer",
    6: "Creative Arts/Entertainment",
    7: "Banking/Consulting/Finance/Marketing/Business/CEO/Entrepreneur/Admin",
    8: "Real Estate",
    9: "International/Humanitarian Affairs",
    10: "Undecided",
    11: "Social Work",
    12: "Speech Pathology",
    13: "Politics",
    14: "Pro sports/Athletics",
    15: "Other",
    16: "Journalism",
    17: "Architecture"
}

length_map = {
    1: "Too little",
	2: "Too much",
	3: "Just Right"
}

numdat_map = {
    1: "Too few",
	2: "Too many",
	3: "Just right"
}

data["gender"] = data["gender"].map(gender_map)
data["condtn"] = data["condtn"].map(condtn_map)
data["match"] = data["match"].map(yes_no_map)
data["samerace"] = data["samerace"].map(yes_no_map)
data["race_o"] = data["race_o"].map(race_map)
data["dec_o"] = data["dec_o"].map(yes_no_map)
#data["met_o"] = data["met_o"].map(yes_no_map) valeurs incoherentes  1 2 3 5 6 7 8
data["field_cd"] = data["field_cd"].map(field_code_map)
data["race"] = data["race"].map(race_map)
data["goal"] = data["goal"].map(goal_map)
data["date"] = data["date"].map(date_go_out_map)
data["go_out"] = data["go_out"].map(date_go_out_map)
data["career_c"] = data["career_c"].map(career_code_map)
data["dec"] = data["dec"].map(yes_no_map)
#data["met"] = data["met"].map(yes_no_map)  valeurs incoherentes 0 1 2 3 5 6 7 8
data["length"] = data["length"].map(length_map)
data["numdat_2"] = data["numdat_2"].map(numdat_map)
data["date_3"] = data["date_3"].map(yes_no_map)

# num_in_3 --> pas compris / pas coherent
```

Distribution de l'âge

```python
age_mean = data_unique["age"].mean()
age_median = data_unique["age"].median()

age = px.histogram(
    data_unique, 
    x="age",
    title="Distribution de l'âge"
)
age.add_vline(
    x=age_mean,
    line_dash='dash',
    annotation_position='top',
    line_color="red",
    annotation=dict(
        text=f'Moyenne ({age_mean.round(1)})',
        font=dict(color='red')
    )
)
age.add_vline(
    x=age_median,
    line_dash='dot',
    annotation_position='bottom',
    line_color="blue",
    annotation=dict(
        text=f'         Médiane ({age_median})',
        font=dict(color='blue')
    )
)
age.update_layout(width=1800, height=600)
age.show()

show_images(age, "../images/age.png")

age_by_sex = px.histogram(
    data_unique, 
    x="age", 
    color="gender", 
    barmode="group", 
    color_discrete_map={
        "Female": "red",
        "Male": "blue"
        },
    title="Distribution de l'âge suivant le sexe"
)
age_by_sex.update_layout(width=1800, height=600)
age_by_sex.show()

show_images(age_by_sex, "../images/age_by_sex.png")
```

Dans le 1er graphique, la distribution des âges révèle une population jeune et concentrée.  
La plupart ont entre 22 et 30 ans.
La moyenne (26,4 ans) et la médiane (26 ans) sont très proches.  
Quelques cas rares dépassent 32 ans (jusqu'à 55 ans).
  
Dans le 2e graphique, on regarde la différence entre les hommes et les femmes.  
Les femmes sont plus représentées avant 27 nas, tandis qu'au-delà, les hommes deviennent majoritaires.  

Distribution de la race

```python
race_data = data_unique["race"].value_counts()
race = px.pie(
    values=race_data, 
    names=race_data.index,
    title="Proportion de la race"
)
race.update_layout(width=1800, height=400)
race.show()

show_images(race, "../images/race.png")
```

Dans ce graphique, on voit que la plupart des participants sont Caucasiens/Européens-Américains (environ 55%), ce qui correspond aux étudiants de l'université Columbia à cette époque (2002-2004).  
Ensuite viennent les Asiatiques/Pacific Islanders/Asiatiques-Americains qui représentent 25% de l'échantillon, ce qui fait d'eux le 2ème groupe le plus important.  
Les autres origines (Noirs américains, Latinos, etc.) sont beaucoup moins nombreuses et restent minoritaires.

Distribution des domaindes étudiés

```python
etude = px.histogram(
    data_unique.sort_values("field_cd"), 
    x="field_cd", 
    title="Distribution des domaines d'étude"

)
etude.update_layout(width=1800, height=400)
etude.show()

show_images(etude, "../images/etude.png")

etude_by_sex = px.histogram(
    data_unique.sort_values("field_cd"), 
    x="field_cd", 
    color="gender", 
    barmode="group", 
    color_discrete_map={
        "Female": "red",
        "Male": "blue"
        },
    title="Distribution des études suivant le sexe"

)
etude_by_sex.update_layout(width=1800, height=400)
etude_by_sex.show()

show_images(etude_by_sex, "../images/etude_by_sex.png")
```

On voit que la plupart étudie commerce/finance, biologie ou ingénierie – ces 3 domaines représentent plus de la moitié des participants.
Ceci correspond aux étudiants de Columbia qui choisissent des études qui mènent à de bons métiers.  
Par contre, très peu étudie les langues ou l'architecture – moins d'une dizaine de personnes.

Dans le 2e histogramme, les hommes choisissent surtout commerce/finance et ingénierie (métiers techniques qui payent bien).  
Les femmes préfèrent sciences sociales/psychologie, enseignement ou travail social (métiers avec les gens).  
C'est une différence classique entre ce que choisissent les femmes et les hommes à l'université à cette époque.
